```python
# Import the pathlib to csv library
from pathlib import Path 
import csv
import os 

# Set the file path
csvpath = Path("budget_data.csv")

# Initialize a list for data
budget_data = [] 

# Initialize vairables for the for loop 
total_months = 0
net_total_profit_loss = 0
previous_profit_loss = 0
profit_loss = 0
changes = {}
line_number = 0

# Open the input path (budget_data) as a file object so it can be used to obtain data
with open(csvpath, 'r') as csvfile:
    
    # Pass in the csv file to the csv.read() function & (with ',' as the delimiter) 
    csvreader = csv.reader(csvfile, delimiter= ',')
    
    # Go to the next row from the start of the file 
    # (which is often the first row/header) and iterate line_number by 1
    header = next(csvreader)
    line_number += 1
    
    # Print header 
  #  print(f"{header} <--- HEADER") UN-HASH TO REVEAL HEADER

# row number
    row_number = 0
    
# 1ST FOR LOOP
    
     # Read each row of data after the header
    for row in csvreader: 
        total_months += 1
        net_total_profit_loss += int(row[1])
        # Print the row
        #print(row)  UN-HASH TO REVEAL DATASET
        row_number += 1
        if row_number > 1:
            previous_profit_loss = profit_loss
            profit_loss = int(row[1])
            changes.update({row[0]:(profit_loss - previous_profit_loss)})
        else:
            previous_profit_loss = 0
            profit_loss = int(row[1])
 

# Vairiables for 

month_greatest_increase = ""

greatest_increase = 0

month_greatest_decrease = ""

greatest_decrease = 0

total_change = 0

for month,change in changes.items():
    total_change += change
    if change > greatest_increase:
        greatest_increase = change
        month_greatest_increase = month
    if change < greatest_decrease:
        greatest_decrease = change
        month_greatest_decrease = month

average_change = round(total_change / len(changes),2) 
# print("\n\n\n") UN-HASH FOR SPACES WHEN PRINTING DATASET


print("Financial Analysis")
print("--------------------------")
print(f"Total months: {total_months}")
print(f"Total: ${net_total_profit_loss}")
print(f"Average Change ${round(average_change,2)}")
print(f"Greatest Increase in Profit {month_greatest_increase}: ${greatest_increase}")
print(f"Grestest Decrease in Profit {month_greatest_decrease}: ${greatest_decrease}")
     
        
    
    
    
  
```
